```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel('http://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx')
```

```python
data.head()
```

```python
data['Description'] = data['Description'].str.strip()
data.dropna(axis=0,subset=['InvoiceNo'], inplace=True)
data['InvoiceNo'] = data['InvoiceNo'].astype('str')
data = data[~data['InvoiceNo'].str.contains('C')]
data.head(10)
```

```python
data['Country'].value_counts()
```

```python
myBasket = (data[data['Country']=='Germany']
       .groupby(['InvoiceNo', 'Description'])['Quantity']
       .sum().unstack().reset_index().fillna(0)
        .set_index('InvoiceNo'))
myBasket.head(10)
```

```python
def my_encode_units(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
myBasketSet = myBasket.applymap(my_encode_units)
myBasketSet.drop('POSTAGE', inplace=True, axis=1)
myBasketSet.head()
```

## Training Model 

```python
my_freq_itemsets = apriori(myBasketSet, min_support=0.07, use_colnames=True)
```

```python
my_rules = association_rules(my_freq_itemsets, metric='lift', min_threshold=1)
my_rules.head()
```
